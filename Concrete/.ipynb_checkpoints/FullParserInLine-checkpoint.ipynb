```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy as sci
  

#Creates function to remove extra spaces in sim file
def cleanlines(line):
    remove = ['  ', ' ']
    for r in remove:
        clean = line.replace(r, '') 
    split = clean.split(';')
    return split
```

```python
#Creates empty list for x, y, z, and energy value of hits
xlist = []
ylist = []
zlist = []
elist = []
lastx = []
lasty = []
lastz = []
```

```python
data = open('FullConcreteModel.Co60_1332.sim', 'r')
```

```python
#Opens sim file
counter = 1
#Starts tallies in detector at 0
c = 0
counts = 0
hold_lines = []
found = False 
for l in data:
    
    line = cleanlines(l.strip('\n'))
    
    if line[0][0:2] == 'ID':  #len(line) == 1 and
        found = True
        hold_lines.append(line)
        c +=1
        continue
    if found == True:
        hold_lines.append(line)
        c +=1
        if line[0][0:2] == 'SE':
            
            hold_lines = []
            found = False
            c=0
            continue
        if line[0][2:6] == 'BLAK':  #len(line) >= 1 and
            counts +=1
            elist.append(np.float32(line[14]))
            xlist.append(np.float32(line[4]))
            ylist.append(np.float32(line[5]))
            zlist.append(np.float32(line[6]))
            
            try:
                #print(hold_lines[c-2])
                lastx.append(np.float32(hold_lines[c-2][4]))
                lasty.append(np.float32(hold_lines[c-2][5]))
                lastz.append(np.float32(hold_lines[c-2][6]))
                c = 0
            except IndexError:
                print('ooo')
                c = 0
                pass
           
            found = False
            hold_lines = []
            
data.close()
```

```python
print(counts)
```

```python
Tot = np.array(list(zip(xlist,ylist,zlist,elist,lastx,lasty,lastz)))
print(Tot.shape)
```

```python
FullE = max(Tot[:,3])
FullE_arr = Tot[Tot[:,3]==FullE] #Creates array of only full energy deposition
Att_arr = Tot[(Tot[:,3]<FullE) & (Tot[:,6]>0)] #Creates array of less than full energy deposition - Attenuated
Scat_arr = Tot[(Tot[:,3]<FullE) & (Tot[:,6]<=0)] #Creates array of less than full energy deposition - Scattered
print(FullE_arr.shape)
print(Att_arr.shape)
print(Scat_arr.shape)
```

```python
FullE_arr[:, 4] =  0
FullE_arr[:, 5] =  0
FullE_arr[:, 6] =  250
```

```python
#Converting to cylindrical coordinates (r,phi,z)

#Generates phi -180 to 180 degrees for Total Detector Response
phi1 = np.arctan2(Tot[:,1],Tot[:,0])*180/np.pi
z1 = Tot[:,2]
#Generates phi -180 to 180 degrees for Full Energy Deposition
phi2 = np.arctan2(FullE_arr[:,1],FullE_arr[:,0])*180/np.pi
z2 = FullE_arr[:,2]
#Generates phi -180 to 180 degrees for Partial Energy Deposition due to Atteunation through Concrete
phi3 = np.arctan2(Att_arr[:,1],Att_arr[:,0])*180/np.pi
z3 = Att_arr[:,2]
#Generates phi -180 to 180 degrees for Partial Energy Deposition due to Scattering off Asphalt
phi4 = np.arctan2(Scat_arr[:,1],Scat_arr[:,0])*180/np.pi
z4 = Scat_arr[:,2]
```

```python
#Confirms constant r of 750 cm for all datapoints 
r = np.sqrt(Tot[:,0]**2+Tot[:,1]**2)
print(r)
```

```python
xedges = np.arange(-180.0, 180.1, 3.)
yedges = np.arange(0., 1000.1, 40.)
heatmap, _, __ = np.histogram2d(phi1,z1, bins=[xedges, yedges])
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

plt.clf()
plt.figure(figsize=(18.8496,8))
im1 = plt.imshow(heatmap.T, extent=extent, aspect='auto', origin='lower')
plt.xticks(np.arange(-180., 181., 30.))
# plt.xlim(100, 180)
cb = plt.colorbar(im1)
cb.set_label('Hits')
plt.title('Total Detector Response = ' + str(len(Tot)) +' hits')
plt.xlabel('Phi (Degrees)')
plt.ylabel('Z (cm)')
plt.savefig('images/1Total_.png')
plt.show()
```
